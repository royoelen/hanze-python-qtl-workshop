```python
import numpy as np

import seaborn as seaborn
import matplotlib.pyplot as plt

import pandas as pandas

# read the genotype data
genotypes = pandas.read_table('/Users/royoelen/hanze-master/2021/CeD_genotypes_adjusted27082018.txt', sep='\t', index_col = 0)
# read the expression data
expression = pandas.read_table('/Users/royoelen/hanze-master/2021/geuvadis_normalised_gene_expression_adjusted27082018.txt', sep='\t', index_col = 0)
```

```python
# extract the gene
expression_gene = expression.loc['ENSG00000179344']
# extract the genotype
genotype_variant = genotypes.loc['rs9273493']
# combine into one dataframe
variant_to_gene = pandas.merge(genotype_variant, expression_gene, right_index = True, left_index = True)
```

```python
# set the size of the plot
f, ax = plt.subplots(figsize=(7, 6))

# now create the plot
seaborn.boxplot(x="rs9273493", y="ENSG00000179344", data=variant_to_gene,
            whis=[0, 100], width=.6, palette="vlag")

# Add in points to show each observation
seaborn.stripplot(x="rs9273493", y="ENSG00000179344", data=variant_to_gene,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
seaborn.despine(trim=True, left=True)
```

```python
# read the positions files
snp_positions = pandas.read_table('/Users/royoelen/hanze-master/2021/snp_locations_CeD_adjusted27082018.txt', sep='\t', index_col = 0)
gene_positions = pandas.read_table('/Users/royoelen/hanze-master/2021/gene_locations.txt', sep='\t', index_col = 0)
```

```python
snp_positions
```

```python
gene_positions
```

```python
# get the genes that are on chromosome 1
gene_positions_chr1 = gene_positions[gene_positions['chr'] == 'chr1'].index.values.tolist()
# and the snps that are on position 1
snp_positions_chr1 = snp_positions[snp_positions['chr'] == 'chr1'].index.values.tolist()
```

```python
# get all the genes
all_genes = expression.index.values.tolist()
# create a list of indices of genes we want to keep
indices_genes_chr1 = []
for index in range(len(all_genes)):
    # add to the index list if for this chromosome
    if all_genes[index] in gene_positions_chr1:
        indices_genes_chr1.append(index)

# subset our genes
expression_chr1 = expression.iloc[indices_genes_chr1]
```

```python
expression_chr1
```

```python
# now for the snps
all_snps = genotypes.index.values.tolist()
# create a list of indices of snps to keep
indices_snps_chr1 = []
for index in range(len(all_snps)):
    # add to the index list if for this chromosome
    if all_snps[index] in snp_positions_chr1:
        indices_snps_chr1.append(index)

# subset our genes
genotypes_chr1 = genotypes.iloc[indices_snps_chr1]
```

```python
genotypes_chr1
```

```python
# write the result
expression_chr1.to_csv('/Users/royoelen/hanze-master/2021/geuvadis_normalised_gene_expression_adjusted27082018_chr1.tsv', sep = '\t')
genotypes_chr1.to_csv('/Users/royoelen/hanze-master/2021/CeD_genotypes_adjusted27082018_chr1.txt', sep = '\t')
```
